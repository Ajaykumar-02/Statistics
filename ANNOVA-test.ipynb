```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df['pclass'].unique()
```

```python
# 3. Data Cleaning: ANOVA cannot handle missing (NaN) values. 
# We filter for only 'age' and 'pclass' and drop rows where age is missing.
df = df[['age','pclass']].dropna()
```

```python
# 4. Grouping the data: Separate the 'age' column into three groups based on 'pclass'
# Class 1: Luxury/First Class, Class 2: Middle Class, Class 3: Lower Class
class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']
class_1
```

```python
# 5. Perform One-Way ANOVA
# Null Hypothesis (H0): The mean age is the same across all classes.
# Alternative Hypothesis (H1): At least one class has a different mean age.
f_stats,p_value = f_oneway(class_1,class_2,class_3)
```

```python
f_stats
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("reject the null hypothesis and there is a signafacant difference in atleast one passenger class")
else:
    print("there is no significant difference")
```
