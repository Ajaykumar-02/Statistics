```python
import numpy as np
```

```python
arr = np.array([1,1,2,2,3,4,5,5,6,6,7,8,9,9,10,10,11,13,13,14,15,101])
```

```python
arr
```

```python
Q1 = np.percentile(arr,25)
Q3 = np.percentile(arr,75)
IQR = Q3 -Q1
IQR
```

```python
Upper_fance = Q3 +(1.5*IQR)
Upper_fance
Lower_fance = Q1 - (1.5*IQR)
Lower_fance
```

```python
l = []
for i in arr:
    if i >=Lower_fance and i<=Upper_fance:
        l.append(i)
arr2 = np.array(l)        
```

```python
print(arr)
print(arr2)
```
