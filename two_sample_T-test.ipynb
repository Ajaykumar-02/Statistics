```python
import numpy as np 
from scipy import stats
```

```python
group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88] 
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83] 
```

```python
# 2. Perform Welch's T-test (Independent Two-Sample T-test)
# We set equal_var=False because we aren't assuming the groups have the same variance.
# This returns the T-statistic (difference magnitude) and the P-value (probability).
t_stats , p_value = stats.ttest_ind(group_A, group_B , equal_var= False)
```

```python
t_stats
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("I will reject the null hypothesis ")
else:
    print("i will accept the null hypothesis" )
```
